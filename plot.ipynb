```python
import csv
import matplotlib.pyplot as plt
```

```python
def plot_curves(name):
    # Open the log file
    # Read the log file and extract the values for epoch, train_loss, and val_loss
    epoch = []
    train_loss = []
    val_loss = []
    with open(f'results/{name}_log_train', 'r') as logfile:
        logreader = csv.reader(logfile, delimiter=',')
        next(logreader)  # Skip the header row
        for row in logreader:
            epoch.append(int(row[0]))
            train_loss.append(float(row[1].split('(')[1].split(',')[0]))
            val_loss.append(float(row[2].split('(')[1].split(',')[0]))

    # Create a figure and plot the train_loss and val_loss curves
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(epoch, train_loss, label='Train Loss')
    ax.plot(epoch, val_loss, label='Val Loss')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.set_title(f'{name} Training Loss vs Validation Loss')
    ax.legend()
    plt.show()
```

```python
name = 'nih_25e'
plot_curves(name)
```

```python
name = 'nih_25_lr_decay033'
plot_curves(name)
```

```python
name = 'chexphoto_25_lr_decay10'
plot_curves(name)
```
